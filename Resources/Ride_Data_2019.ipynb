```python
import pandas as pd
```

```python
# load the files
jan = "jan2019data.csv"
feb = "feb2019data.csv"
mar = "mar2019data.csv"
apr = "apr2019data.csv"
may = "may2019data.csv"
jun = "jun2019data.csv"
jul = "jul2019data.csv"
aug = "aug2019data.csv"
sept = "sept2019data.csv"
octb = "oct2019data.csv"
nov = "nov2019data.csv"
dec = "dec2019data.csv"

# Read files and store into Pandas data frame
jan_rides = pd.read_csv(jan)
feb_rides = pd.read_csv(feb)
mar_rides = pd.read_csv(mar)
apr_rides = pd.read_csv(apr)
may_rides = pd.read_csv(may)
jun_rides = pd.read_csv(jun)
jul_rides = pd.read_csv(jul)
aug_rides = pd.read_csv(aug)
sept_rides = pd.read_csv(sept)
octb_rides = pd.read_csv(octb)
nov_rides = pd.read_csv(nov)
dec_rides = pd.read_csv(dec)
```

```python
jan_rides.head()
```

```python
# Using DataFrame.insert() to add a month column 
jan_rides.insert(1, "Month", "January", True) 
feb_rides.insert(1, "Month", "February", True) 
mar_rides.insert(1, "Month", "March", True) 
apr_rides.insert(1, "Month", "April", True) 
may_rides.insert(1, "Month", "May", True) 
jun_rides.insert(1, "Month", "June", True) 
jul_rides.insert(1, "Month", "July", True) 
aug_rides.insert(1, "Month", "August", True) 
sept_rides.insert(1, "Month", "September", True) 
octb_rides.insert(1, "Month", "October", True) 
nov_rides.insert(1, "Month", "November", True) 
dec_rides.insert(1, "Month", "December", True) 
  
# Observe the result 
jan_rides.head()
```

```python
#combine the csv files
total_2019 = pd.concat([jan_rides, feb_rides,mar_rides,apr_rides,may_rides,
                        jun_rides,jul_rides,aug_rides,sept_rides,octb_rides,nov_rides,dec_rides ],ignore_index=True,sort=True)
```

```python
total_2019.head()
```

```python
#convert the gender column from numeric to string (0 =unknown gender, 1 = male, 2 = female)
total_2019.loc[total_2019["gender"] == 0, "gender"] = "unknown"
total_2019.loc[total_2019["gender"] == 1, "gender"] = "male"
total_2019.loc[total_2019["gender"] == 2, "gender"] = "female"
```

```python
total_2019.head()
```

```python
#get age by taking the current year minus birth year
total_2019["Age"] = pd.Series([])
```

```python
total_2019.head()
```

```python
total_2019["Age"] = 2019 - total_2019['birth year']
```

```python
total_2019.head()
```

```python
# Create bins for ages
age_bins = [0, 9, 14, 19, 24, 29, 34, 39, 44, 49, 54, 59, 64, 69, 74, 79, 84, 89, 94, 99, 300]

# Create labels for these bins
group_labels_age = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34",
                "35-39", "40-44", '45-49', '50-54', '55-59', '60-64', '65-69', '70-74',
                   '75-79', '80-84', '85-89', '90-94', '95-99', '100+']
    
    
# Slice the data and place it into bins and do calculations
total_2019['Age Group'] =  pd.cut(total_2019["Age"], age_bins, labels=group_labels_age)
```

```python
total_2019.head()
```

```python
total_2019 = total_2019.rename(columns={'tripduration': "trip duration (sec)"})
total_2019.head()
```

```python
total_2019["trip duration (min)"] = total_2019['trip duration (sec)'] / 60
total_2019.head()
```

```python
total_2019.to_csv('Total_2019_Rides.csv')
```
